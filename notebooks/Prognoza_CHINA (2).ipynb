```python
df["year_centered"] = df["year"] - df["year"].min()
df["year_squared"] = df["year_centered"] ** 2

adf_result_c = adfuller(df["ml_models"], regression="c", autolag="AIC")  # Bez trendu
adf_result_ct = adfuller(df["ml_models"], regression="ct", autolag="AIC")  # Z trendem liniowym

X_quad = sm.add_constant(df[["year_centered", "year_squared"]])
model_quad = sm.OLS(df["ml_models"], X_quad).fit()
residuals_quad = model_quad.resid
adf_result_quad = adfuller(residuals_quad, regression="c", autolag="AIC")

test_results_adf = {
    "ADF (linear trend)": {
        "Statistic": adf_result_ct[0],
        "p-value": adf_result_ct[1],
        "Lags": adf_result_ct[2],
        "Critical Values": adf_result_ct[4],
    },
    "ADF (quadratic trend)": {
        "Statistic": adf_result_quad[0],
        "p-value": adf_result_quad[1],
        "Lags": adf_result_quad[2],
        "Critical Values": adf_result_quad[4],
    },
}

test_results_adf
```

### Model Holta w podziale na zbiór uczący i testowy z różnicowaniem

```python
# 1. Różnicowanie szeregu czasowego

diff_series = df['ml_models'].diff().dropna()
diff_years = df['year'][1:]  # Dopasowanie lat do zróżnicowanego szeregu

# 2. Podział na zbiór uczący i testowy (ostatnie 4 lata jako test)
train_size_diff = len(diff_series) - 3
train_diff, test_diff = diff_series[:train_size_diff], diff_series[train_size_diff:]
train_years_diff = diff_years[:train_size_diff]
test_years_diff = diff_years[train_size_diff:]

# 3. Dopasowanie modelu Holta-Wintersa do zróżnicowanego szeregu
model_diff = ExponentialSmoothing(train_diff, trend='add', seasonal=None)
fit_diff = model_diff.fit()

# 4. Prognoza na 4 lata (na zróżnicowanym szeregu)
forecast_diff = fit_diff.forecast(steps=3)

# 5. Odróżnicowanie prognozy (dodajemy ostatnią wartość ze zbioru uczącego)
last_actual_value = df['ml_models'].iloc[train_size_diff]  # Ostatnia rzeczywista wartość przed prognozą
forecast_final = last_actual_value + forecast_diff.cumsum()  # Suma skumulowana różnic

# 6. Obliczenie błędów
actual_values = df['ml_models'][train_size_diff + 1:].values  # Rzeczywiste wartości dla 4 lat testowych
predicted_values = forecast_final.values

plt.figure(figsize=(14, 6))
plt.plot(train_years_hw, train_hw, label='Oryginalne dane', marker='o')
plt.plot(test_years_hw, test_hw, label='Rzeczywiste dane (zbiór testowy)', marker='o', linestyle='--')
plt.plot(test_years_diff, predicted_values, label='Prognoza Holta-Wintersa (po różnicowaniu)', color='red', linestyle='--', marker='x')
plt.xlabel("Rok")
plt.ylabel("Liczba modeli ML")
plt.title("Prognoza na 4 lata (po różnicowaniu szeregu)")
plt.legend()
plt.grid(True)
plt.show()
```
